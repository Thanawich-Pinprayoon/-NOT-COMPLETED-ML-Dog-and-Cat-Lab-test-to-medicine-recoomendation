```
class Case:
  def __init__(self,id = None):
    self.id = id
    # if id = None:
    #   self.id = -1
    # else:
    #   self.id = id

    # if list == None:
    self.list = []
    for _list in drug_list:
      self._list = None
    for __i in drug_list:
      self.list.append("NA")
  def makeList(self):
    list = []
    for _list in drug_list:
      list.append(self._list)
  def set_id(self,id):
    self.id = id
  
  def __str__(self):
    return str(self.id)
```

```

def getCol(drugName):
  drugName = str(drugName).lower()
  if drugName[-1] == " ":
    drugName = drugName.strip()
  if drugName == 'amoxicillin':
    drugName = 'amoxicillin/clavulanic'
  if drugName == 'amoxicillin/clavulanic acid':
    drugName = 'amoxicillin/clavulanic'
  elif drugName == 'piperacillin-tazobactam':
    drugName = 'piperacillin'
  elif drugName ==  'polymixin b':
    drugName = 'polymyxin b'
  elif drugName == 'amoxy/clav':
    drugName = 'amoxicillin/clavulanic'
  elif drugName == 'cephalothin':
    drugName = 'cefalotin'
  elif drugName == 'cefalothin':
    drugName = 'cefalotin'
  elif drugName == 'oxacillin*':
    drugName = 'oxacillin'
  elif drugName == 'minocyclilne':
    drugName = 'minocycline'
  elif drugName == 'benzyl penicillin':
    drugName = 'benzylpenicillin'
  elif drugName == 'cephalothin':
    drugName = 'cefalotin'
  elif drugName == 'inducible clindamycin resistance':
    drugName = 'inducible clindamycin'
  elif drugName == 'cephalexin':
    drugName = 'cefalexin'
  elif drugName == 'rifampin':
    drugName = 'rifampicin'
    
  for index_drugList in range(len(drug_list)):
    # print(drugName,drug_list[index_drugList])
    if drugName == drug_list[index_drugList]:
      return int(index_drugList)
   
  print('================================================================')
  print('      ERROR : drug name : .'+str(drugName)+'. NOT FOUND!')
  print('================================================================')
  if drugName not in drug_not_found:
    drug_not_found.append(drugName)
  return None
```

```
cases = []
ids_list = []
drug_not_found = []
for index_Lab in range(len(Lab_df['id'])):
  # print(index_Lab)
  #dropped 0
  if index_Lab == 0:
    continue
  
  #make class
  this_id = int(Lab_df['id'].loc[index_Lab])
  #print('id = ',this_id,end=" ")
  if this_id not in ids_list:
    while this_id != len(cases)-1:
      cases.append(Case(index_Lab))
    # print("cases size = ", len(cases))
    ids_list.append(this_id)
    # print(' this id = ',this_id)

  #get drug name
  drug_name = Lab_df['anti_name'].loc[index_Lab]
  # print('name : ',drug_name)

  #find col of the drug
  sel_col = getCol(drug_name)
  if sel_col == None:
    # break
    continue
  # print('col = ',sel_col)

  #put result
  # print('   make list ',index_Lab)
  cases[this_id].makeList()
  cases[this_id].list[sel_col] = Lab_df['sir'].loc[index_Lab]
  # print('   -end make list ',index_Lab)
  # print(len(cases[this_id].list))
  # print("==============================")
```
